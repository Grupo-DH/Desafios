2. Limpeza e transformação da variável alvo **Salário Mensal**: verifique o tipo de dado da variával Salário Mensal, faça as devidas mudanças, plote a distribuição e dispersão dos dados e aplique a transformação matemática para a correção da assimetria.

```python
# Verificando o tipo da variável salário mensal
df['Salário Mensal'].dtypes
```

```python
# Plotando a distribuição da variável salário mensal
plt.hist(df['Salário Mensal'], bins=100, range = (1, 25000), label="Salario Mensal")
plt.legend();
```

```python
# Plotando o gráfico de dispersão da variável salário mensal
plt.scatter(x=df['Salário Mensal'].index,y=df['Salário Mensal'],alpha=0.5)
plt.show()
```

```python
# Inserindo o logaritimo no df e verificando seu histograma
df['log_salario_mensal'] = np.log1p(df['Salário Mensal'])
plt.hist(df['log_salario_mensal'], 
         bins=50,
         label="log Salario Mensal")
plt.legend();
```

```python
#Normalizando o log_salario_normal
mean = df['log_salario_mensal'].mean()
std = df['log_salario_mensal'].std()
df['normal_log_salario_mensal'] = (df['log_salario_mensal']-mean)/std
plt.hist(df['normal_log_salario_mensal'], 
         bins=50,
         label="Normal log Salario Mensal")
plt.legend();
```

```python
# Executando o teste de normalidade de shapiro, cseguindo média 0 e variância 1
resultado_normal = stats.shapiro(df['normal_log_salario_mensal'])
resultado_normal
```

3. Verificação dos dados da variável Idade: verifique os dados da variável Idade, plote uma distribuição e uma dispersão, calcule a correlação com a variável salário mensal e interprete os resultados.

Verificação dos dados da variável **Sexo**: verifique os dados da variável Sexo, plote uma distribuição e uma dispersão, calcule a correlação com a variável salário mensal e interprete os resultados. Depois disso, faça uma inferência estatística pra verificar se os homens ganham mais que as mulheres.

```python
df['Sexo'].dtypes
```

```python
salario_mulheres = df.query('Sexo == 2')
salario_homens = df.query('Sexo == 1')
fig, ax = plt.subplots(1,1)
plt.hist(salario_mulheres['normal_log_salario_mensal'], bins = 20, label = "Salario Mensal - Mulheres", histtype = 'step')
plt.hist(salario_homens['normal_log_salario_mensal'], bins = 20, label = "Salario mensal - Homens", histtype = 'step')
plt.legend(loc='upper left', frameon =False);
```

```python
fig, ax = plt.subplots(1,1)
plt.scatter(x = salario_homens['Salário Mensal'].index, y = salario_homens['Salário Mensal'], label="salario homens", alpha = 0.5)
plt.scatter(x = salario_mulheres['Salário Mensal'].index, y = salario_mulheres['Salário Mensal'], label="Salario mulheres", alpha = 0.5)
plt.legend();
```

```python
stats.ttest_ind(a=df.normal_log_salario_mensal[df.Sexo == 1],
                b=df.normal_log_salario_mensal[df.Sexo == 2], 
                equal_var = False)
```

6. Verificação dos dados da variável **Raça Cor**: verifique os dados da variável Raça Cor, plote uma distribuição e uma dispersão, calcule a correlação com a variável salário mensal e interprete os resultados. Depois disso, faça uma inferência estatística pra verificar se os brancos ganham mais que os outros grupos étnicos.

```python
df['Raça Cor'].dtypes
```

```python
fig, ax = plt.subplots(1,1)
plt.hist(df['normal_log_salario_mensal'][df['Raça Cor'] == 2], bins = 20, label = 'Salário Mensal - brancos', histtype = 'step')
plt.legend(loc='upper left', frameon =False);
```

```python
fig, ax = plt.subplots(1,1)
plt.hist(df['normal_log_salario_mensal'][df['Raça Cor'] != 2], bins = 20, label = 'Salário mensal - outros grupos', histtype = 'step')
plt.legend(loc='upper left', frameon =False);
```

```python
# Plotando um gráfico de dispersão para o salário mensal de brancos e outros grupos etnicos.
fig, ax = plt.subplots(1,1)
plt.scatter(x = df['Salário Mensal'][df['Raça Cor'] == 2].index, y = df['Salário Mensal'][df['Raça Cor'] == 2], label="salario brancos", alpha = 0.5)
plt.scatter(x = df['Salário Mensal'][df['Raça Cor'] != 2].index, y = df['Salário Mensal'][df['Raça Cor'] != 2], label="Salario outros grupos", alpha = 0.5)
plt.legend();
```

```python
# aplicando o t-test para médias de duas séries
stats.ttest_ind(a=df['normal_log_salario_mensal'][df['Raça Cor'] == 2],
                b=df['normal_log_salario_mensal'][df['Raça Cor'] != 2], 
                equal_var = False)
```

```python
# aplicando o t-test para médias de duas séries
stats.ttest_ind(a=df['normal_log_salario_mensal'][df['Raça Cor'] == 2],
                b=df['normal_log_salario_mensal'][df['Raça Cor'] == 4], 
                equal_var = False)
```

8. Avançado: plote um SEABORN heatmap com todas as correlações das variávies numéricas entre si.

```python
# Montando um DataFrame com apenas as variáveis analisadas
df2 = pd.DataFrame(df['Salário Mensal'])
df2['Idade'] = df['Idade']
df2['Grau Instrução'] = df['Grau Instrução']
df2['Sexo'] = df['Sexo']
df2['Raça Cor'] = df['Raça Cor']
df2.head()
```

```python
sns.heatmap(df2.corr(), annot = True, fmt= '.2g', square = True, vmin = 0, vmax = 1)
```
